```python
from plyer import notification
from bs4 import BeautifulSoup
import time
import smtplib
from plyer import notification
```

```python
def notifyME(title,message):
    notification.notify(
        title=title,
        message=message,
        app_icon=None,
        timeout = 5
    )
```

```python
import requests
r=requests.get('https://zeenews.india.com/')
a=r.text
soup=BeautifulSoup(a,'html.parser')
soup
y=soup.find(class_='coroupdatetable')
y
z=y.find_all('tr')
z
l=[]
for i in range(len(z)):
   
    l.append(z[i].get_text())
for i in range(len(l)):
    l[i]=l[i].strip("\n")
    l[i]=l[i].strip()
#     print(l[i].strip())    
# print(z[0].get_text().strip("\n").strip())
z[0].get_text().strip("\n").strip()
x="har"
sl=[]
for i in l:
    if x.lower() in i.lower():
        sl.append(i)
# for k in sl:
#      print("".join(k))
# sl            
# z=y.find_all('tr')
# z
for i in range(len(sl)):
    sl[i]=sl[i].split("\n\n")
    sl[i].insert(1,"\n\n Total case")
    sl[i].insert(3,"\n\n Death")
    print("".join(sl[i]))        
    notifyME("covid 19","".join(sl[i]))
    time.sleep(6)
```
